```python
import random
import pandas as pd

class BancoDeDados:
    def __init__(self):
        self.usuarios = pd.DataFrame(columns=['usuario', 'senha', 'pontos', 'interacoes'])

    def adicionar_usuario(self, usuario, senha):
        if not self.usuarios['usuario'].str.contains(usuario).any():
            novo_usuario = pd.DataFrame({'usuario': [usuario], 'senha': [senha], 'pontos': [0], 'interacoes': [0]})
            self.usuarios = pd.concat([self.usuarios, novo_usuario], ignore_index=True)
            print(f"Usuário {usuario} registrado com sucesso!")
        else:
            print(f"Usuário {usuario} já está registrado.")

    def validar_login(self, usuario, senha):
        usuario_existente = self.usuarios[self.usuarios['usuario'] == usuario]
        if not usuario_existente.empty:
            if usuario_existente['senha'].values[0] == senha:
                return True
        return False

    def interagir(self, usuario):
        interacao = random.choice(["Quiz sobre Fórmula E", "Enquete sobre corridas", "Compartilhar em mídias sociais"])
        print(f"\nInteração disponível: {interacao}")
        
        escolha = input("Deseja participar? (s/n): ").strip().lower()
        if escolha == 's':
            pontos_ganhos = random.randint(100, 500)
            self.usuarios.loc[self.usuarios['usuario'] == usuario, 'pontos'] += pontos_ganhos
            self.usuarios.loc[self.usuarios['usuario'] == usuario, 'interacoes'] += 1
            print(f"Você ganhou {pontos_ganhos} pontos participando do {interacao}!")
        else:
            print("Você escolheu não participar da interação.")

    def colocar_pontos(self, usuario, pontos):
        if self.usuarios['usuario'].str.contains(usuario).any():
            self.usuarios.loc[self.usuarios['usuario'] == usuario, 'pontos'] += pontos
            print(f"{pontos} pontos adicionados ao usuário {usuario}.")
        else:
            print(f"Usuário {usuario} não encontrado.")

    def exibir_ranking(self):
        ranking = self.usuarios[['usuario', 'pontos', 'interacoes']].sort_values(by='pontos', ascending=False)
        print("\nRanking de Fãs da Fórmula E:")
        print(ranking)

def login(usuario, senha, banco):
    if banco.validar_login(usuario, senha):
        print("\nLogin bem-sucedido!")
        return True
    else:
        print("\nUsuário ou senha incorretos.")
        return False

banco_dados = BancoDeDados()

while True:
    print("\nBem-vindo ao sistema de promoção da Fórmula E!")
    print("1. Registrar novo usuário")
    print("2. Fazer login")
    print("3. Sair")
    escolha = input("Escolha uma opção (1, 2 ou 3): ")

    if escolha == '1':
        usuario_input = input("Digite o nome de usuário para registro: ")
        senha_input = input("Digite sua senha: ")
        banco_dados.adicionar_usuario(usuario_input, senha_input)
    
    elif escolha == '2':
        usuario_input = input("Digite seu nome de usuário: ")
        senha_input = input("Digite sua senha: ")

        if login(usuario_input, senha_input, banco_dados):
            while True:
                print("\n1. Interagir com conteúdo da Fórmula E")
                print("2. Ver ranking de fãs")
                print("3. Logout")
                opcao = input("Escolha uma opção (1, 2 ou 3): ")

                if opcao == '1':
                    banco_dados.interagir(usuario_input)

                elif opcao == '2':
                    banco_dados.exibir_ranking()

                elif opcao == '3':
                    print("Você fez logout com sucesso!")
                    break
                else:
                    print("Opção inválida, tente novamente.")

    elif escolha == '3':
        print("Saindo do sistema. Até mais!")
        break

    else:
        print("Opção inválida. Tente novamente.")
```
